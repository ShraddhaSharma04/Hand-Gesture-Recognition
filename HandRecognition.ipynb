```python
# First, install required packages
!pip install mediapipe opencv-python numpy

import cv2
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import output
import time

class HandGestureRecognizer:
    def __init__(self):
        self.mp_hands = mp.solutions.hands
        self.hands = self.mp_hands.Hands(
            static_image_mode=False,
            max_num_hands=2,
            min_detection_confidence=0.7,
            min_tracking_confidence=0.5
        )
        self.mp_draw = mp.solutions.drawing_utils

    def detect_gesture(self, landmarks):
        """
        Detect gestures based on finger positions
        Returns the name of the gesture
        """
        # Get finger tip and pip landmarks
        thumb_tip = landmarks[4]
        index_tip = landmarks[8]
        middle_tip = landmarks[12]
        ring_tip = landmarks[16]
        pinky_tip = landmarks[20]

        # Get corresponding pip (middle finger joints)
        thumb_pip = landmarks[3]
        index_pip = landmarks[6]
        middle_pip = landmarks[10]
        ring_pip = landmarks[14]
        pinky_pip = landmarks[18]

        # Check if fingers are up or down
        fingers_up = []

        # Special check for thumb
        if thumb_tip.x < thumb_pip.x:
            fingers_up.append(1)
        else:
            fingers_up.append(0)

        # Check for other fingers
        for tip, pip in [(index_tip, index_pip),
                        (middle_tip, middle_pip),
                        (ring_tip, ring_pip),
                        (pinky_tip, pinky_pip)]:
            if tip.y < pip.y:
                fingers_up.append(1)
            else:
                fingers_up.append(0)

        # Recognize gestures
        if sum(fingers_up) == 0:
            return "Fist"
        elif sum(fingers_up) == 5:
            return "Open Hand"
        elif fingers_up == [0, 1, 0, 0, 0]:
            return "Index Point"
        elif fingers_up == [0, 1, 1, 0, 0]:
            return "Peace Sign"
        elif fingers_up == [1, 1, 0, 0, 1]:
            return "Spider-Man"
        else:
            return "Unknown Gesture"

    def process_frame(self, frame):
        # Convert BGR to RGB
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Process the frame
        results = self.hands.process(rgb_frame)

        # Draw hand landmarks
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                # Draw landmarks
                self.mp_draw.draw_landmarks(
                    frame,
                    hand_landmarks,
                    self.mp_hands.HAND_CONNECTIONS
                )

                # Get landmarks as list
                landmarks = []
                for landmark in hand_landmarks.landmark:
                    landmarks.append(landmark)

                # Detect gesture
                gesture = self.detect_gesture(landmarks)

                # Put gesture text on frame
                cv2.putText(
                    frame,
                    gesture,
                    (10, 50),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1,
                    (0, 255, 0),
                    2
                )

        return frame

    def run_webcam(self):
        """
        Run detection on webcam feed
        Note: This needs to be modified for Google Colab
        """
        cap = cv2.VideoCapture(0)

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            frame = self.process_frame(frame)
            cv2_imshow(frame)
            output.clear()

            # Break loop on 'q' press
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()

# Create instance and run
recognizer = HandGestureRecognizer()
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode, b64encode
import cv2
import numpy as np
import PIL
import io
import html
import time
```

```python
import cv2
import mediapipe as mp
import numpy as np
import time

class HandGestureRecognizer:
    def __init__(self):
        self.mp_hands = mp.solutions.hands
        self.hands = self.mp_hands.Hands(
            static_image_mode=False,
            max_num_hands=2,
            min_detection_confidence=0.7,
            min_tracking_confidence=0.5
        )
        self.mp_draw = mp.solutions.drawing_utils

    def detect_gesture(self, landmarks):
        thumb_tip = landmarks[4]
        index_tip = landmarks[8]
        middle_tip = landmarks[12]
        ring_tip = landmarks[16]
        pinky_tip = landmarks[20]

        thumb_pip = landmarks[3]
        index_pip = landmarks[6]
        middle_pip = landmarks[10]
        ring_pip = landmarks[14]
        pinky_pip = landmarks[18]

        fingers_up = []

        # Thumb check
        if thumb_tip.x < thumb_pip.x:
            fingers_up.append(1)
        else:
            fingers_up.append(0)

        # Other fingers
        for tip, pip in [(index_tip, index_pip),
                        (middle_tip, middle_pip),
                        (ring_tip, ring_pip),
                        (pinky_tip, pinky_pip)]:
            if tip.y < pip.y:
                fingers_up.append(1)
            else:
                fingers_up.append(0)

        # Gesture recognition
        if sum(fingers_up) == 0:
            return "Fist"
        elif sum(fingers_up) == 5:
            return "Open Hand"
        elif fingers_up == [0, 1, 0, 0, 0]:
            return "Index Point"
        elif fingers_up == [0, 1, 1, 0, 0]:
            return "Peace Sign"
        elif fingers_up == [1, 1, 0, 0, 1]:
            return "Spider-Man"
        else:
            return "Unknown Gesture"

    def start_camera(self):
        # Initialize the webcam
        cap = cv2.VideoCapture(0)  # 0 is usually the default laptop camera

        # Set camera resolution
        cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
        cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

        # FPS calculation variables
        prev_time = 0
        curr_time = 0

        while True:
            # Read frame from webcam
            success, frame = cap.read()
            if not success:
                print("Failed to read from camera")
                break

            # Flip the frame horizontally for a later selfie-view display
            frame = cv2.flip(frame, 1)

            # Convert BGR to RGB
            rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            # Process the frame and detect hands
            results = self.hands.process(rgb_frame)

            # Calculate FPS
            curr_time = time.time()
            fps = 1 / (curr_time - prev_time)
            prev_time = curr_time

            # Display FPS
            cv2.putText(frame, f'FPS: {int(fps)}', (10, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

            # Draw hand landmarks and detect gestures
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    # Draw landmarks
                    self.mp_draw.draw_landmarks(
                        frame, hand_landmarks, self.mp_hands.HAND_CONNECTIONS,
                        self.mp_draw.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                        self.mp_draw.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2)
                    )

                    # Get landmarks list
                    landmarks = []
                    for landmark in hand_landmarks.landmark:
                        landmarks.append(landmark)

                    # Detect and display gesture
                    gesture = self.detect_gesture(landmarks)
                    cv2.putText(frame, f'Gesture: {gesture}', (10, 70),
                              cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

            # Show the frame
            cv2.imshow('Hand Gesture Recognition', frame)

            # Break loop on 'q' press
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        # Release the webcam and destroy windows
        cap.release()
        cv2.destroyAllWindows()

# Create and run the recognizer
if __name__ == "__main__":
    recognizer = HandGestureRecognizer()
    recognizer.start_camera()
```
